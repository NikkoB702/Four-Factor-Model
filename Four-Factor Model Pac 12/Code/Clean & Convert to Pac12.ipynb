```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "C:/Users/Home/College_Basketball_Dataset/cbb.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
df['CONF'].value_counts()
```

```python
# Filtering the data down to PAC12 teams
cbb_teams = df[ (df.CONF == 'P12')].copy()
cbb_teams
```

```python
cbb_teams[['TEAM']].describe()
```

```python
# Removing null values specific to "Postseason" and "Seed" and replacing with "Missed Tournament"

cbb_teams.fillna({'POSTSEASON':'Missed Tournament'}, inplace = True)
cbb_teams.POSTSEASON.value_counts(dropna=False)
```

```python
cbb_teams.fillna({'SEED':'Missed Tournament'}, inplace = True)
cbb_teams.SEED.value_counts(dropna=False)
```
